# Application of Convolution


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

# Create a signal

```python
Srate = 256 # in Hz (if sampling rate decrease --> siganl will be less smooth )
t = np.arange(0, 3, 1/Srate)
print(t)
```

# Create a noiseless signal

```python

x = np.sin( 2 * np.pi * 2 *  t )
print(x)
```

# Create a random noise

```python
noise = np.random.randn(len(t))
print(noise)
```

# Linear noise need to Incorpate: we need to add it with signal amplitude to get Noisy signal

```python
NoisySignal = x + noise
```

# Draw Original signal

```python
plt.plot(t, x , color="green" )
plt.suptitle("Original-Noiseless signal")
plt.draw()
```

# Draw as 
* stem: discreate time drawing (no color parameter)
* Plot : Continuous time drawing

```python
plt.stem(t, x)
plt.suptitle("Original-Noiseless signal")
plt.draw()
```

# Draw Noisy signal

```python
# color range ( 0 to 1)
plt.plot(t, NoisySignal , color=(1,0,0) )
plt.suptitle("Noisy signal")
plt.draw()
```

# Design a basic media filter/kernel/impulse

```python
filter = np.ones(50)/10.0
print(filter)
```

# Get Filtered signal

```python
filtered_signal = np.convolve(NoisySignal, filter, mode='full')
# print(filtered_signal)
```

```python
plt.plot(filtered_signal , color=(1,0,0) ) # has both X and Y--> in covolved signal
plt.suptitle("Filtered signal")
plt.draw()
```
